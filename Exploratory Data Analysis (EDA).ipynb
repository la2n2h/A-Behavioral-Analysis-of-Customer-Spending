### Approach to Customer Segmentation
segment customers based on key factors such as:

Age,
Gender,
Income Level,
Occupation,
Region,
Membership Type,
The goal is to identify potential customers and loyal customers.

```python
# Visualize distribution
columns = ['Age', 'Gender', 'Income Level', 'Occupation', 'Region', 'Membership Type']
for col in columns:
    sns.countplot(data=df, x=col, palette='coolwarm', order = df[col].value_counts().index)
    plt.title(f'Customer Distribution by {col}')
    plt.xticks(rotation=90)
    plt.show()
```

1. Customer Distribution by Age: 
there are fluctuations in customer numbers across different age groups.
Some age groups have peaks, indicating that certain age ranges contain more customers.
This could mean that specific age groups are more engaged with the product/service, possibly due to lifestyle preferences or purchasing power.

2. Customer Distribution by Gender: 
The gender ratio is relatively balanced between male and female customers.
The number of female customers is slightly higher than male customers.

3. Customer Distribution by Income Level:
The majority of customers fall into the middle-income category.
Low and high-income levels have similar numbers, but the high-income group is slightly smaller.
This may indicate that the market primarily targets the middle class.

4. Customer Distribution by Occupation:
The distribution of occupations is relatively even.
"Doctor," "Artist," and "Engineer" have the highest numbers, suggesting they are key customer segments.
"Teacher" has fewer customers compared to other professions, possibly due to lower demand for the product/service.

5. Customer Distribution by Region:
Most customers are from the "Northeast," "West," and "Midwest" regions.
The "South" region has significantly fewer customers.
This could be due to the market not being as developed in that area or lower demand.


7. Customer Distribution by Membership Type
The "VIP" membership has the highest number of customers, indicating a strong preference for premium membership.
"New Customer" and "Regular" have nearly equal numbers, suggesting a stable influx of new customers.
The high "VIP" membership count may reflect the success of the loyalty program.


Conclusion:
Customer distribution is uneven across demographic factors, with fluctuations in age groups rather than a steady decline. While younger customers still make up a significant portion, there are notable peaks in specific age groups, suggesting varying engagement levels across different life stages.
The majority of customers have a middle income and come from the Northeast, West, and Midwest regions. Occupation also influences distribution, with doctors, artists, and engineers being the primary customer groups. The high number of VIP members indicates a successful loyalty program, appealing to a broad customer base across age group.

###  Identify factors that influence spending patterns

```python
categories = ["Occupation", "Membership Type", "Income Level"]
for col in categories:
    sns.boxplot(data=df, x=col, y="Total Amount", palette="coolwarm", order=df[col].value_counts().index)
    plt.title(f"Total Spending by {col}")
    plt.xticks(rotation=45)
    plt.show()
```

### comparision between occoupations' distribution and their spending
###### Student & Teacher:
Customer Count: Among the smallest groups in the dataset.
Spending Behavior:
Although the number of students and teachers is low, their spending is relatively high.
Their spending distribution has a wide range, meaning there are individuals who spend significantly more than others.
Outliers exist, indicating that some students and teachers are high-value customers.
Insight:
This group may have specific needs or purchasing habits that justify higher spending.
Marketing strategies could focus on offering exclusive deals for students and teachers to increase retention and engagement.
###### Artist:
Customer Count: One of the larger occupation groups.
Spending Behavior:
Despite having a large number of customers, artists tend to spend less compared to other groups.
Their median spending is among the lowest.
The presence of multiple outliers suggests that while most artists spend conservatively, a few spend significantly more.
Insight:
Artists might be more budget-conscious or have irregular income, leading to conservative spending.
However, targeting high-spending artists with personalized offers or exclusive product categories might be effective.
###### Doctor & Engineer:
Customer Count: Among the highest in number.
Spending Behavior:
Both groups spend significantly more than other occupations.
Their spending range is wide, meaning some doctors and engineers are heavy spenders.
They have high-value outliers, indicating that premium products or services may be attractive to them.
Insight:
This group likely has higher disposable income and is willing to spend more.
Luxury items, premium memberships, or high-end product recommendations could be targeted toward them.
###### Business Analyst & Salesperson:
Customer Count: Moderate representation.
Spending Behavior:
Their spending patterns are balanced, with neither too many low spenders nor extreme outliers.
Business Analysts tend to spend slightly more than Salespersons.
Insight:
Since they show stable spending, marketing strategies should focus on upselling and increasing purchase frequency rather than trying to attract new customers in this group.
###### Lawyer:
Customer Count: Moderate.
Spending Behavior:
Spending is on the lower side with fewer high-value outliers.
Most lawyers have a conservative spending pattern compared to other professions.
Insight:
Lawyers may be cautious in their spending habits, so marketing strategies should focus on value-for-money deals rather than premium offerings.
